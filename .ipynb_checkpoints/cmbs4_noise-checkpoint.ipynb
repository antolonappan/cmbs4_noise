```python
import os
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
import pysm3
import pysm3.units as u
from fgbuster import (CMB, Dust, Synchrotron,
                      basic_comp_sep,harmonic_ilc,
                      get_observation, get_instrument, MixingMatrix)

from noise import S4_LAT
```

```python
NSIDE=256
sky = pysm3.Sky(nside=NSIDE, preset_strings=["c1"])
```

```python
noise = S4_LAT()
Nl_T= noise.noise_curves_T()
Nl_P= noise.noise_curves_P()
```

```python
def ilc_weights(m_list):
    y = np.vstack(m_list)
    C = np.cov(y)
    C_inv = np.linalg.inv(C)
    w = np.sum(C_inv, axis=1) / np.sum(C_inv)
    return w
def cs_map(m_list, w):
    m = sum(a * b for a, b in zip(w, m_list))
    return m
```

```python
bands = noise.bands
beams = noise.beams
maps_list = []
for idx in range(len(bands)):
    print(idx)
    maps = sky.get_emission(bands[idx]* u.GHz)
    maps= maps.to(u.uK_CMB, equivalencies=u.cmb_equivalencies(bands[idx]*u.GHz))
    Nl_T[idx][0] = 0
    noiseI = hp.synfast(Nl_T[idx],nside=NSIDE)
    maps_list.append(maps.value[0]+noiseI)
```

```python
w = ilc_weights(maps_list)
cmb = cs_map(maps_list,w)
```

```python
hp.mollview(cmb)
```

```python
noise_list = []
for idx in range(len(bands)):
    print(idx)
    Nl_T[idx][0] = 0
    noiseI = hp.synfast(Nl_T[idx],nside=NSIDE)
    noise_list.append(noiseI)
```

```python
plt.figure(figsize=(8,8))
for idx in range(len(bands)):
    plt.loglog(hp.alm2cl(hp.map2alm(noise_list[idx])),label=f"{bands[idx]}")
plt.loglog(hp.alm2cl(hp.map2alm(cs_map(noise_list,w))),c='k',label="coadded")
plt.legend(fontsize=15)
plt.xlabel("$\ell$",fontsize=20)
plt.ylabel("$N_\ell^T$",fontsize=20)
```
